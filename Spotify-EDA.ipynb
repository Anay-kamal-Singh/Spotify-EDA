## Artist with most danceability song

```python
most_danceability_songs = df[["danceability", "song_title", "artist"]].sort_values(by="danceability",ascending=False)[:5]
```

```python
most_danceability_songs
```

```python
plt.figure(figsize=(12,7))
sns.barplot(x="danceability", y="artist", data=most_danceability_songs)
plt.title("Artist with most danceability song")
plt.show()
```

## Top 10 instrumentalness Tracks

```python
instrumental_tracks = df[["instrumentalness", "song_title", "artist"]].sort_values(by="instrumentalness",ascending=False)[:5]
```

```python
instrumental_tracks
```

```python
plt.figure(figsize=(12,7))
plt.pie(x='instrumentalness', data=instrumental_tracks, autopct='%1.2f%%', labels=instrumental_tracks.song_title)
plt.show()
```

## Multiple feature plots

```python
interest_feature_cols = ['danceability']
```

```python
for feature_col in interest_feature_cols:
    pos_data = df[df["target"]==1][feature_cols]
    neg_data = df[df["target"]==0][feature_cols]
    
    plt.figure(figsize=(12,7))
    
    sns.histplot(pos_data, bins=30, label="Positive", color="green")
    sns.histplot(neg_data, bins=30, label="Negative", color="red")
    
    plt.title(f"Positive And Negative Histogram Plot For {feature_cols}")
    plt.show()
```
